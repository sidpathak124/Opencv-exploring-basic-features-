### HSV (Hue, Saturation and Value) 

HSV separates image luminance from color information, 
hence is used where color description is very important

HSV are represented by a 3d cylindrical cone, H = along circumference, S = towards the radius, V = Along the length

Hue: correponds to color components(base pigments)                       range = 0-360, RYGCBM
Saturation: The amount of color (depth of the pigment, dominance of hue) range = 0-100%
Value: Brightness of the color                                           range = 0-100%

Process: Comparing each pixel of an image with a predefined threshold value. Such comparison divides all the pixels (of i/p image) into two groups; Px with intensities above the threshold and lower than the threshold

### Morphological transformations

<b>What are they?</b><br>
Morphological transformations are simple transformations on the image shape. They are normally performed on binary image.<br>
<b>What is required?</b><br>
1) The original image<br>
2) A structuring element (AKA kernel) which decides the nature of an operation.<br>
ex: Changing the value of each pixel based on the avg. value of neighboring pixels<br>
**Note:** M.T are normally done on grayscale images

```python
import cv2 as cv
%matplotlib qt5
smarties = cv.imread('smarties.png', cv.IMREAD_GRAYSCALE)

_, mask1 = cv.threshold(smarties, 220, 255, cv.THRESH_BINARY_INV)

j = cv.imread('j.png', 0) 

# applying a kernel on an image
kernel = np.ones((2,2), np.uint8) # Bigger the kernel lesser the discontinuities, however this also expands the object
dilation = cv.dilate(mask1, kernel, iterations = 2) # More the iterations lesser the discontinuities 
erosion = cv.erode(mask1, kernel, iterations = 5) # erodes away boundary of the foreground object
opening = cv.morphologyEx(mask1, cv.MORPH_OPEN, kernel) # erosion and then dilation
closing = cv.morphologyEx(mask1, cv.MORPH_CLOSE, kernel)# dilation and then erosion
mg = cv.morphologyEx(mask1, cv.MORPH_GRADIENT, kernel)  # morphological gradient = dilation - erosion
th = cv.morphologyEx(mask1, cv.MORPH_TOPHAT, kernel)    # tophat = image - opening

images = [smarties, mask1, dilation, erosion, opening, closing, mg, th]
title = ['smarties', 'mask1', 'dilation', 'erosion', 'opening', 'closing', 'mg', 'th']

for i in range(8):
    plt.subplot(2, 4, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(title[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```

## Image gradients and canny edge detection

**Image gradient** is a directional change in the intensity or color in an image. It is the fundamental building block of image processing. Use Case: for finding edges in an image.

Here we discuss 3 Image gradient methods
1) laplacian method: uses laplacian derivatives
both uses joint gaussian and differentiation operations
2) sobelx method: 
3) sobely method  

```python
import cv2 as cv
%matplotlib tk

image = cv.imread('sudoku.png', 0)
# LAP DETECTS EDGES
lap = cv.Laplacian(image, cv.CV_64F, ksize = 3) # dtype = 64float, to store negative nums generated by laplacian
lap = np.uint8(np.absolute(lap)) # CONVERTING TO UINT8
sobelx = cv.Sobel(image, cv.CV_64F, 1, 0) # dx = 1 (order of derivative in the x-direction), dy = 0
sobelx = np.uint8(np.absolute(sobelx))    # change in intensity is in the x-direction (detects change in y-direction)
sobely = cv.Sobel(image, cv.CV_64F, 0, 1) # dx = 0 , dy = 1 (order of derivative in the x-direction)
sobely = np.uint8(np.absolute(sobely))    # change in intensity is in the y-direction (detects change in x-direction)
# combining sobelx and sobely
sobelxy = np.bitwise_or(sobelx, sobely)

images = [image, lap, sobelx, sobely, sobelxy]
titles = ['image', 'lap', 'sobelx', 'sobely', 'sobelxy']

for i in range(5):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```

### Canny edge detector
An edge detection operator that uses a multi stage algorithm to detect a wide range of images<br>
It is composed of 5 steps -:<br>
1) **Noise reduction:**             Apply gaussian filter<br>
2) **Gradient Calculation:**        Find intensity gradients of the image<br>
3) **Non-maximum suppression:**     Get rid of spurious response to edge detection<br>
4) **Double threshold**             Determine potential edges<br> 
5) **Edge tracking by hysterisis**  Finalize detection of edges by suppressing all other edges that are weak or not connected to                                   strong edges

**Note** This technique results in a lot less noise as compared to the above used gradient techniques

```python
import cv2 as cv
%matplotlib tk

sudoku = cv.imread('sudoku.jpg', 0)
cv.imshow('Original', sudoku)

cv.namedWindow('Canny_detection')
# cv.imshow('image', image)
def nothing(x):
    pass

cv.createTrackbar('Upper_th', 'Canny_detection', 100, 200, nothing) # nothing = A call back fn
cv.createTrackbar('Lower_th', 'Canny_detection', 0, 100, nothing)


while 1:
    
    
    k = cv.waitKey(1) & 0xFF
    if k == 27: # 27 == escape key
        break
        
    u_th = cv.getTrackbarPos('Upper_th', 'Canny_detection')
    l_th = cv.getTrackbarPos('Lower_th', 'Canny_detection')

    canny = cv.Canny(sudoku, l_th, u_th)
    cv.imshow('Canny_detection', canny)
        
cv.destroyAllWindows()
```

## Image Pyramids

Sometimes we need to work with an image of different resolutions. We search for <br>
our desired object in all of these different resolutions of the same image.<br>

**Pyramid** is a type of multiscale signal representation in which a signal or an image<br>
is subject to repeated smoothing and subsampling.<br>

**Ex.** Original Image **---->** level_1 (1/2 resolution) + smoothing and subsampling **---->** level_2 (1/4 resolution) + smoothing and subsampling .... and so on....

There are two types of pyramids available in cv -:<br>
1) Gaussian pyramid: repeat filtering and resampling of an image<br>
        a) Pyrdown
        b) Pyrup
2) Laplacian pyramid

```python
import cv2 as cv
%matplotlib tk

image = cv.imread('lena.png', -1)
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
lr1 = cv.pyrDown(image)
lr2 = cv.pyrDown(lr1)
hr1 = cv.pyrUp(lr2)

images = [lr1, lr2, hr1]
titles = ['lr1', 'lr2', 'hr1']

for i in range(3):
    
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()
```

Laplacian pyramid is formed with the help of two techniques of gaussian pyramid (pyrup and pyrdown), there is no separate method for gaussian. 

A level in Laplacian pyramid is formed by difference between that level in Gaussian pyramid and extended level of its upper level in Gaussian pyramid

**Usefulness:**<br> 
Gaussian and Laplacian pyramids helps us to blend the images and reconstruction of images

```python
import cv2 as cv

image = cv.imread('lena.png', -1)
layer = image.copy()
gp = [layer]

for i in range(6):
    layer = cv.pyrDown(layer)
    gp.append(layer)
    
upper_layer = gp[5]

for i in range(5, 0, -1):
    extended_layer = cv.pyrUp(gp[i])
    laplacian = cv.subtract(gp[i-1], extended_layer)
    cv.imshow(f'laplayer {i}', laplacian)

cv.waitKey(0)
cv.destroyAllWindows()
```

## Image blending using pyramids

The 5 steps to follow -:<br>
1. Load the two images
2. Find Gaussian pyramid for the two images
3. From Gaussian pyramid find their Laplacian pyramid
4. Join the left half with the right half in each levels of Laplacian pyramid
5. From these joint image pyramids, reconstruct the original image

```python
# Performing only an image join

import cv2 as cv

apple = cv.imread('apple.jpg', -1)
orange = cv.imread('orange.jpg', -1)
# mat stack method
mix = np.hstack((apple[:, :256], orange[:, 256:]))

# cv.imshow('apple', apple)
# cv.imshow('orange', orange)
cv.imshow('mix', mix)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Performing image blending

import cv2 as cv

apple = cv.imread('apple.jpg', -1)
orange = cv.imread('orange.jpg', -1)

# mat stack method
mix = np.hstack((apple[:, :256], orange[:, 256:]))

apple_copy = apple.copy()
gp_apple = [apple_copy]

# Generate gp for apple
for i in range(6):
    apple_copy = cv.pyrDown(apple_copy)
    gp_apple.append(apple_copy)
    
orange_copy = orange.copy()
gp_orange = [orange_copy]

# Generate gp for orange
for i in range(6):
    orange_copy = cv.pyrDown(orange_copy)
    gp_orange.append(orange_copy)
    
# Generate laplacian for apple
apple_copy = gp_apple[5]
lp_apple = [apple_copy]

for i in range(5,0,-1):
    gaussian_extended = cv.pyrUp(gp_apple[i])
    laplacian = cv.subtract(gp_apple[i-1], gaussian_extended)
    lp_apple.append(laplacian)
    
# Generate laplacian for orange
orange_copy = gp_orange[5]
lp_orange = [orange_copy]

for i in range(5,0,-1):
    gaussian_extended = cv.pyrUp(gp_orange[i])
    laplacian = cv.subtract(gp_orange[i-1], gaussian_extended)
    lp_orange.append(laplacian)

# Adding left and right halves for images on each level
apple_orange_pyramid = []
n = 0
for apple_lap, orange_lap in zip(lp_apple, lp_orange):
    n += 1
    rows, cols, ch = apple_lap.shape
    laplacian = np.hstack((apple_lap[:, :int(cols/2)], orange_lap[:, int(cols/2):]))
    apple_orange_pyramid.append(laplacian)
    
# Reconstructing from all the joined laplacians
apple_orange_reconstruct = apple_orange_pyramid[0]

for i in range(1,6):
    apple_orange_reconstruct = cv.pyrUp(apple_orange_reconstruct)
    apple_orange_reconstruct = cv.add(apple_orange_pyramid[i], apple_orange_reconstruct)
    
cv.imshow("mix", mix)
cv.imshow("apple_orange_reconstruct", apple_orange_reconstruct)
    
cv.waitKey(0)
cv.destroyAllWindows()
```

### Find and draw Contours 

**What are contours?**<br>
The curve joining all the continuous points along the boundary which are having the same color or intensity

**Useful**<br>
For shape analysis, object detection, object recognition

For better accuracy ,binary images are generally used for finding the contours

```python
import cv2 as cv

image = cv.imread('opencv_logo.png')
imggray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(imggray, 150, 255, 0)
# RETR_TREE = retrieval tree, 
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_NONE)
# countours = A list of all contours in the image. Each contour is a np array of (x,y) 
# coordinates of boundary points of the object.
# hierarchy contains info. about image topology
# print("Number of contours = {}".format(len(contours)))
# print(contours[0])
 
cv.drawContours(image, contours, -1, (255, 0, 0), 3)
    
cv.imshow('image', image)
cv.imshow('thresh', thresh)

cv.waitKey(0)
cv.destroyAllWindows()
```

### Basic motion detection 

```python
import cv2 as cv
import numpy as np

cap = cv.VideoCapture('vtest.avi')

ret, frame1 = cap.read()
ret, frame2 = cap.read()

while cap.isOpened():
    
    diff = cv.absdiff(frame1, frame2)
    gray = cv.cvtColor(diff, cv.COLOR_BGR2GRAY)
    blur = cv.GaussianBlur(gray, (5,5), 0)
    _, thresh = cv.threshold(blur, 20, 255, cv.THRESH_BINARY)
    dilated = cv.dilate(thresh, None, iterations = 3)
    contours, _ = cv.findContours(dilated, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
    
    for contour in contours:
        (x,y,w,h) = cv.boundingRect(contour)
        
        if cv.contourArea(contour) < 700:
            continue
        cv.rectangle(frame1, (x,y), (x+w,y+h), (0,255,0), 2)
        cv.putText(frame1, "status: Movement", (10, 20), cv.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 3)
            
#     cv.drawContours(frame1, contours, -1, (0, 255, 0), 2)
    
    cv.imshow('feed', frame1)
    frame1 = frame2
    ret, frame2 = cap.read()
        
    if cv.waitKey(40) == 27:
        break
        
cv.destroyAllWindows()
cap.release()
```
